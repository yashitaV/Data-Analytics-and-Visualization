## Ques-1 : Central limit theorem

### Uniform Distribution

```python
## GENERATING UNIFORM IID RANDOM VARIABLES
U = np.random.uniform(0,1,1000)
```

```python
### Creating 1000 samples with sample size of 25 for Uniform distribution ###
## Making dataframe to visulaize in form of table ##

N = 25
df = pd.DataFrame()
for i in range(1,1001):
    sample = np.random.choice(U,size = 25)
    sample = sample - sample.mean()
    col = f'X {i}'
    df[col] = sample   
df.head(3)
```

```python
## Cumulative sum ##

cum_sum = df.transpose().cumsum()
cum_sum[:24].transpose().plot(kind='density',/
                subplots = True,layout = (5,5),figsize = (8,8))
```

### Normal distribution

```python
### Normal random distribution with mean at 0 and std = 1 ###
Nor = np.random.normal(0,1,1000)
```

```python
### Creating 1000 samples with sample size of 25 for Normaldistribution ###
N = 25
N_df = pd.DataFrame()

for i in range(1,1001):
    sample_norm = np.random.choice(Nor,25)
    sample_norm = sample_norm - sample_norm.mean()
    col = f'sample {i}'
    N_df[col] = sample_norm   

N_df.head(3)    
```

```python
## Cumulative sum ##

cum_sum = N_df.transpose().cumsum(axis = 0, skipna = True)
cum_sum[:24].transpose().plot(kind='density',subplots = True,/
                              layout = (5,5),figsize = (8,8))
```

### Laplacian Distribution

```python
### laplace random distribution with mean at 0 and lambda = 1 ###
Lap = np.random.laplace(0,1,1000)
```

```python
### Creating 1000 samples with sample size of 25 for Laplacian distribution ###
N = 25
Lap_df = pd.DataFrame()

for i in range(1,1001):
    sample_Lap = np.random.choice(Lap,size = 25)
    sample_Lap = sample_Lap - sample_Lap.mean()
    col = f'X {i}'
    Lap_df[col] = sample_Lap    

Lap_df.head(3) 
```

```python
## Cumulative sum ##

cum_sum = Lap_df.transpose().cumsum(axis = 0, skipna = True)
cum_sum[:24].transpose().plot(kind='density',subplots = True,/
                              layout = (5,5),figsize = (8,8))
```

#### Conclusions:
The cumulative sum of the distribution (irrespective of the type of distribution) converged to Normal distribution, Hence
verifying the Central limit theorem

## Ques-3 : Sampling distribution of the difference between two means

###### (A) For P1(â„¦) âˆ¼ N(5,1) and P2(â„¦) âˆ¼ N(3,1) and for n1 < 30 and n2 < 30.

```python
P1 = np.random.normal(5,1,100)
P2 = np.random.normal(3,1,100)
```

```python
mean_diff_df = pd.DataFrame()

for i in range(1000):
    diff = []
    P1_sample = np.random.choice(P1, size=25)
    P2_sample = np.random.choice(P2, size=25)
    diff.append(P1_sample.mean() - P2_sample.mean())
    col = f'X {i}'
    mean_diff_df[col] = diff  

mean_diff_df.head(5)
```

```python
mean_diff_df.transpose().plot(kind = 'density')
```

```python
print('Mean of difference of sample means is: ', mean_diff_df.transpose().mean())
print('Normal Approximation mean :', P1.mean()-P2.mean() )
print('Variance of difference of sample means is: ', mean_diff_df.transpose().var())
print('Normal Approximation variance :', (P1.var()/(25) + P2.var()/(25)) )
```
