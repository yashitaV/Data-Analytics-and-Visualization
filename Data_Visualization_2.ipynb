### 1. Cereal Dataset

```python
Cer= pd.read_csv('cereal.csv')
```

```python
Cer.head()
```

```python
                    #### VILOIN PLOT #####
    
fig = plt.figure(figsize=(14,6))
gs = fig.add_gridspec(1,3)

ax = fig.add_subplot(gs[0, 0])
sns.violinplot(data = Cer['rating'])      
ax.set_title("Rating")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(data = Cer['cups'],palette="Set2")
ax.set_title("Cups")

fig.tight_layout()
plt.show()
```

```python
                            ### QQ PLOT ###

sm.qqplot(Cer.calories, line ='45')
```

```python
## Standardized

scipy.stats.probplot(Cer.calories, dist="norm", plot=plt)
```

```python
                                ### MOSAIC PLOT ###
    
mosaic(Cer, ['mfr','type'])
```
